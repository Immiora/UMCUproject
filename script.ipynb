```python
import scipy.io 
import pandas as pd
import matplotlib.pyplot as plt
import os


mat = scipy.io.loadmat('project/data/Data/F1/mat/usctimit_ema_f1_001_005.mat')
data = mat['usctimit_ema_f1_001_005']
UL_df = pd.DataFrame.from_dict(data[0][1][2])
LL_df = pd.DataFrame.from_dict(data[0][2][2])
JAW_df = pd.DataFrame.from_dict(data[0][3][2])
TD_df = pd.DataFrame.from_dict(data[0][4][2])
TB_df = pd.DataFrame.from_dict(data[0][5][2])
TT_df = pd.DataFrame.from_dict(data[0][6][2])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(UL_df[0], UL_df[1], UL_df[2], color='white', edgecolors='grey', alpha=0.5)
ax.scatter(UL_df[0], UL_df[1], UL_df[2], c='red')
plt.show()
```
