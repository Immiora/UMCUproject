```python
import scipy.io 
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_file = scipy.io.loadmat('data/Data/F1/mat/usctimit_ema_f1_001_005.mat')
data = data_file['usctimit_ema_f1_001_005']
audio_data = pd.DataFrame.from_dict(data[0][0][2])
```

```python
# plot of the first sentence
# "this was easy for us"
first_sent = audio_data.loc[(audio_data.index >= 18000) & (audio_data.index <= 43000)]
ax = first_sent.plot()
```

```python
# plot of the second sentence
# "jane may earn more money by working hard"
second_sent = audio_data.loc[(audio_data.index >= 70000) & (audio_data.index <= 130000)]
ax = second_sent.plot()
```
